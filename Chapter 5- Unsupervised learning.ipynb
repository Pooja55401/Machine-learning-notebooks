## Introduction — Unsupervised Learning

An unsupervised learning is the techinique to draw references from datasets which contains input data without labeled responses. Unsupervised machine learning algorithms tries to find out the patterns from a unlabeled or uncategorised dataset. Unlabled dataset means there is no sample dataset for which outcome is known. Unsupervised machine learning methods cannot be directly applied to any regression or a classification problem. it is not possible to train the algorithm the for any particular problem as in the case of unsupervised learning we don't have any clue about the values for the output. In the area of artifical intelligence, when we are not provided data with output label to classify the data, called as an unsupervised learning problem.
Unsupervised learning can instead be used for discovering the underlying structure of the data. Unsupervised learning can be more unpredictable than supervised learning. 

In the case of unsupervised learning the model learns through the process of observation and tries find out the structures in the given dataset. Once a dataset is feeded to the model, it finds patterns and relationships between the samples in the dataset by creating clusters. Unlike supervised learning, there is no teacher to train the model. Without the guidance of a teacher the model becomes unable to find the hidden structure in unlabeled data by it-self.

## K-Means clustering Algorithm 

K-Means is very well known "clustering" algorithms. K-means algorithm assigns $k$ centroids to the entires data to define clusters. Data points are considered to be a part of a particular cluster if it is closer to the centroid of that particular cluster than the centoried by other clusters. K-Means algorithm tries to find out the best centroids by iteration of 

1. By assigning data points to clusters based on the current centroids 
2. chosing centroids (points which are the center of a cluster) based on the current assignment of data points to clusters. 

Suppose that we have given a training data set ${x^{(1)}, ... , x^{(n)}}$, for each data point $x^{(i)}$ is a feature vectors without any labels $y^{(i)}$. Our aim is to find $k$ clusters where $k<n$ for each datapoint. 
 

Given k, the K-means algorithm works as follows: 

1. Choose k (random) points (centers) to be the initial centroids, cluster centers 
2. Assign each data point to the closest centroid 
3. Re-compute the centroids using the current cluster's mean 
4. If for each cluster the distance between current centoried and mean of the cluster is not equal to zero, repeat steps 2 and 3

Convergence critrion for K-means algortithm is :
If there is no further re-assignments of data points to different clusters $C_j$, or very mintue or no changes in centroids $m_j$ of the cluster $C_j$, or very minute variation in the sum of squared error $||x-m_j||^2$ where, $||x-m_j||$ is the distance between data point $x$ and centroid $m_j$ and also known as Euclidean distance between two points. The objective of K-Means clustering algorithm is to minimize total intra-cluster variance, the squared error. Which is represted as:

$$J=\sum_{j=0}^k \sum_{i=1}^n ||x^{j}_i-c_j||^2$$
where  k is the number of clusters, n is the number of data points, $c_j$ is the centeroid of cluster $j$

K-Means is an quite efficient method for clustering. As number of clusers are specified in advance and the final results are sensitive to initial values of centeroids.  Very often the K-means converges at a local optimum (as shown below) and it does not guarantee to find the global optimum. This condition may depend on the initial clusters. A very practical way is multiple runs of the K-means, which randomly initilize the centorids multiple time (in general 50 to 1000 times), as it is usaually a fast algorithm. After compasring all the values of cost $J$, the cluster with lowest cost can be selected.
 
Compare the cost(J) of multiple runs. In general, a large k probably decreases the error but increases the risk of overfitting.

In K-means algorithm when the assignments no longer change, it automatically converges. The convergence can be stopped by using a different distance function such as spherical k-means and k-medoids that have been proposed to allow other distance measures other than Ecludine distance.

If number of centroids k and dimentio of the plane d are fixed, the K-means can be exactly solved in time $ {\displaystyle O(n^{dk+1})} $, where n is the number of datapoints  to be clustured. 

```python
import numpy as np 
import pandas as pd 
from copy import deepcopy
from matplotlib import pyplot as plt
```

```python
# Set three centers, the model should predict similar results
c1 = np.array([.5,1])
c2 = np.array([4.5,5])
c3 = np.array([7,1])

# Generate random data and center it to the three centers
group1 = np.random.randn(250, 2) + c1
group2 = np.random.randn(250,2) + c2
group3 = np.random.randn(250,2) + c3

data = np.concatenate((group1, group2, group3), axis = 0)

plt.scatter(data[:,0], data[:,1], s=6)
#plt.show()
```

```python
# Number of clusters
k = 3
# Number of training data
n = data.shape[0]
# Number of features in the data
m = data.shape[1]

# Generate random centers, here we use sigma and mean to ensure it represent the whole data
mean = np.mean(data, axis = 0)
std = np.std(data, axis = 0)
centers = np.random.randn(k,m)*std + mean

# Plot the data and the centers generated as random
plt.scatter(data[:,0], data[:,1], s=7)
plt.scatter(centers[:,0], centers[:,1], marker='o', c='r', s=150)
plt.show()
```

```python
centers_old = np.zeros(centers.shape) # to store old centers
centers_new = deepcopy(centers) # Store new centers

data.shape
clusters = np.zeros(n)
distances = np.zeros((n,k))

error = np.linalg.norm(centers_new - centers_old)

# When, after an update, the estimate of that center stays the same, exit loop
while error != 0:
    # Measure the distance to every center
    for i in range(k):
        distances[:,i] = np.linalg.norm(data - centers[i], axis=1)
    # Assign all training data to closest center
    clusters = np.argmin(distances, axis = 1)
    
    centers_old = deepcopy(centers_new)
    # Calculate mean for every cluster and update the center
    for i in range(k):
        centers_new[i] = np.mean(data[clusters == i], axis=0)
    error = np.linalg.norm(centers_new - centers_old)
centers_new    
```

```python
# Plot the data and the centers generated as random
plt.scatter(data[:,0], data[:,1], s=6)
plt.scatter(centers_new[:,0], centers_new[:,1], marker='o', c='r', s=150)
plt.show()
```

### Elbow Method

Elbow method gives us an idea for defining the number of clusters in the dataset. It is based on the sum of squared distance (SSE) matrix between data points and their assigned clusters’ centroids. K has to be picked at the point where SSE starts to flatten out and forming an elbow like shape. For the above generated data set we can eleluate SSE for different values of k and can observe where the curve might form an elbow type shape.


```python
from sklearn.cluster import KMeans
# calculate distortion for a range of number of cluster
SSE = []

for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=0
    )

    km.fit(data)

    SSE.append(km.inertia_)

# plot
plt.plot(range(1, 11), SSE, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

## DBSCAN 

## why?

Partitioning methods such as K-means (and hierarchical clustering) work well for finding spherical or convex shaped clusters. They are commanly used for dense and well-separated clusters. Moreover, they are also very senstive towards noise/outliers present in the data.

DBSCAN algorithm requires two parameters–

1. $\epsilon$- It is used to identify the neighborhood data points that are situated on $\epsilon$ distance from a particular data point. If $\epsilon$ value is too small then majority portions of the data will be considered as noise or outliers. If it is too large then majority of the data points will be in the same clusters. 
One way to find the $\epsilon$ value is based on the k-distance graph. 

2. MinPts: It is the minimum number of data points within $\epsilon$ radius from a choosen data point. The minimum no of MinPts can be determined from the number of dimensions d in the dataset as $MinPts >= d+1$. The minimum value of MinPts must be chosen at least 3. If the dataset is large, the larger value of MinPts has to be chosen.

The $\epsilon$ neighborhood of a data point P in the database DS is defined as //(following the definition from Ester et.al.)
$N(p) = {q ∈ DS | dist(p, q) ≤ ϵ}$

Density-based clustering algorithms tries to cluster a region  where there are lots of points, surrounded by a region where there are few data points. Data points can be divided into three types  :

1. Core Point: Data point $x_i$ is called a core point if it contains more than MinPts points within $\epsilon$ radius.
2. Border Point: Data point which has less than MinPts in its $\epsilon$ radius, but it is in the neighborhood of a core point is called a border point.
3. Noise or outlier: Data point which is not a core point or border point is called outlier.

One problem that can arise with this approach is $\epsilon$ neighborhood of the border points usually contain only very few  points than the core points. If the value of MinPts parameter has choosen low to include the border points in the cluster, then it can cause problem to eliminate the noise. Hence the concept of density-reachable and density-connected points, comes inot the picture.

1. DBSCAN algorithm makes a start with an random starting data point which  has not been visited yet. The neighborhood of this point is identified by using a distance measure epsilon ε. All points which are within the radius ε are considered neighboring points.

2. If there are a minimum data points(minPoints) within its neighborhood then the DBSCAN clustering process begins. The ranodom starting data point becomes the first point in the cluster. 

    Otherwise, the point will be labeled as noise (later this noisy point might become the part of the cluster). In both cases     that point is marked as “visited”.

3. The points within the ε radius also becomes the part of the same cluster and marked as “visited”. For all of the new neighborhoos data points that have been just added to the cluster, the same procedure of including all points in the neighborhood in the same cluster is then repeated.

4. Repeat the steps 2 and 3  until all points within the ε neighborhood of the cluster have been marked as visited.

Once the current cluster has been completely formed, a new unvisited point from the dataset has been retrieved and processed.It  It detemines further cluster or noise (points that doesnt come in $\epsilon$ neighborhood). This process repeats until all points are marked as visited.

DBSCAN poses some great advantages over other clustering algorithms. Firstly, it does not require a pe-set number of clusters at all. It also identifies outliers as noises unlike mean-shift which simply throws them into a cluster even if the data point is very different. Additionally, it is able to find arbitrarily sized and arbitrarily shaped clusters quite well.

## Hierarchical Clustering

Hierarchical clustering is one of the unsupervised machine learning algorithm, which is used to cluster unlabeled data points. Similar to K-means clustering algorithm, hierarchical clustering also groups the data points together based on similarities. We can find that in some cases the outputs of hierarchical and K-Means clustering can be similar. In K-means we have to predefined the number of clusters, but hierarchical clustering give good idea about the number of clusters based on dendograms. Hierarchical clustering find a hierarchy of clusters, where this created hierarchy resembles a tree like structure from data similarities, which called a dendrogram. 

There are two types of hierarchical clustering. 
1. Agglomerative: This is a "bottom-up" approach. In this each data point starts in its own cluster, and pairs of clusters are merged as one moves up the hierarchy. 
2. Divisive: This is a "top-down" approach. Entire dataset start in one cluster, and splits are performed recursively as one moves down the hierarchy.

## Measures of distance (similarity) and linkage criteria

The distance between two clusters has been computed. It is based on length of a straight line drawn from one cluster to another cluster. Which is commonly called as the Euclidean distance. Once the distance metric has choosen, it is necessary to identify from where distance has to be computed. 
For instance, it can be computed between the two most similar parts of a cluster (single-linkage), the two least similar bits of a cluster (complete-linkage), the center of the clusters (mean or average-linkage), or some other criterion.

1. MIN: It is also known as single linkage algorithm. In simple words, pick the two closest data points such that one point lies in one cluster and the other point lies in  another cluster and take their similarity . It can be defined as the similarity of two clusters C1 and C2 is equal to the minimum of the similarity between points Pi and Pj such that Pi belongs to C1 and Pj belongs to C2. Mathematically it can be represented as,

Sim(C1,C2) = Min Sim(Pi,Pj) such that Pi ∈ C1 & Pj ∈ C2

This approach can separate non-elliptical shapes as long as the gap between two clusters is not small. This approach cannot separate clusters properly if there is noise between clusters.

2. MAX: MAX is also known as the complete linkage algorithm. The similarity of two clusters C1 and C2 is equal to the maximum of the similarity between points Pi and Pj such that Pi belongs to C1 and Pj belongs to C2. Mathematically it can be represented as,

Sim(C1,C2) = Max Sim(Pi,Pj) such that Pi ∈ C1 & Pj ∈ C2

This approach does well in separating clusters if there is noise between clusters. Perhapes, this approach is biased towards globular clusters and approach tends to break large clusters.

3. Group Average: In group average, take all the pairs of data points on both the clusters and compute their similarities and calculate the average of the similarities. Mathematically it can be represented as,

sim(C1,C2) = ∑ sim(Pi, Pj)/|C1|*|C2|    where, Pi ∈ C1 & Pj ∈ C2

The group Average approach performs well in separating clusters if there is noise between clusters. This approach is biased towards globular clusters.

4. Distance between centroids: In this approach centroids of two clusters C1 & C2 has ti be calcuated and  similarity between the two centroids is taken as the similarity between two clusters. In practice this is not a popular technique.

5. Ward’s Method: This approach of calculats the similarity between two clusters exactly in the same manner as the  Group Average except that Ward’s method calculates the sum of the square of the distances between data points Pi and PJ, which belongs to the cluster C1 and C2 respectively. Ward method is actually a method that tries to minimize the variance within each cluster.  Mathematically it can be represented as,

sim(C1,C2) = ∑ (dist(Pi, Pj))²/|C1|*|C2|

Ward’s method approach also does well in separating clusters if there is noise between clusters. And is also biased towards globular clusters.

##  Procedure for Agglomerative clustering

Following are the steps involved in agglomerative clustering:

1. In the begining, assume each data point as one cluster. Therefore, the number of clusters at the start will be $N$, $N$ is an integer representing the number of data points in the dataset. 
2. Form a cluster by joining the two most closest data points based on any of the above mentioned similarity measures and then we will left with $N-1$ clusters. 
3. Form more clusters by joining the two most closest data points or clusters based on similarity measure, it will result in $N-2$ clusters. 
4. Repeat the step 2 and 3 until one big cluster is formed. 

Once a single cluster is formed, dendrograms are used to divide into multiple clusters depending upon the problem. We will study the concept of dendrogram in detail in an upcoming section.

## Procedure for Divisive clustering

Its a top down approach
1. Assume entire data as one combined cluster.
2. The cluster splits into two distinct parts or clustures, according to some degree of similarity mentioned above.
3. Clusters split into two again and again until each cluster only contain a single data point.
Divisive clustering is very rarely used.

## Drawbacks of Hierarchical clustering

First of all there is no mathematical objective for Hierarchical clustering. Apart from this, as we have seen in our above discussion that all the approaches involves in  calculating the similarity between clusters has its own certain disadvantages. Hierarchical clustering requires high space and time complexity. The space required for the Hierarchical clustering is very high when the number of data points are high as we need to store the similarity matrix in the RAM. The space complexity is O(n²)
Hence this clustering algorithm cannot be used when we have huge data.

As we have to perform n iterations and in each iteration, we need to update the similarity matrix and restore the matrix, the time complexity is also very high. Time complexity is O(n³) where n is the number of data points. Also, merging of data points or clusters can’t be reversed, which can create a problem if data is noisy, high-dimensional data.
